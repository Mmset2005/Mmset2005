```python
#FranceIOI-lv2-Fonctions-Formes_creuses
def ligne(x,txt):
   res = ""
   for i in range(x):
      res += str(txt)
   print(res)

def carre(y,x):
   ligne(x,'#')
# Parce que sinon y'a deux'#' au lieu d'un...
   if x == 1:
      for i in range(y-2):
         print('#')
# ...et c'est pas ouf
   else:
      for i in range(y-2):
         print('#' + ' '* (x-2) + '#')
   ligne(x,'#')
 
def triangle(x):
   print('@')
   for i in range(x-2):
      print('@' + ' '*i + '@')
   ligne(x,'@')

ligne(int(input()),'X')
carre(int(input()),int(input()))
triangle(int(input()))
```
